```python
img_rgb.shape
```

```python
img = cv2.resize(img_rgb,(1300,275))
```

```python
plt.imshow(img)
```

```python
w_ratio = 0.5
h_ratio = 0.1
```

```python
new_img = cv2.resize(img_rgb,(0,0),img_rgb,w_ratio,h_ratio)
```

```python
plt.imshow(new_img)
```

```python
new_img.shape
```

```python
# Along the horizontal access
new_img = cv2.flip(img_rgb,0)
plt.imshow(new_img)
```

```python
# Flip along vertical access
new_img = cv2.flip(img_rgb,1)
plt.imshow(new_img)
```

```python
# Flip along both axes
new_img = cv2.flip(img_rgb,-1)
plt.imshow(new_img)
```

```python
type(new_img)
```

```python
cv2.imwrite('my_new_picture.jpg',new_img)
```
