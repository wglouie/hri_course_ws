```python
import cv2
import matplotlib.pyplot as plt 
import numpy as numpy
%matplotlib inline
```

```python
img = cv2.imread('data/puppy.jpg')
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img = cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
plt.imshow(img)
```
